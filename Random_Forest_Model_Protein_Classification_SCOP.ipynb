In this Jupyter notebook, we will be building a classification model for the protein dihedral angle data set using the random forest algorithm.

## 1. Import libraries

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

import numpy as np
import pandas as pd
```

## 2. Load the data set

Creating header:

```python
names = []

for i in range(100):
    names.append("phi_" + str(i));
    names.append("psi_" + str(i));

names.append("class")

print(names)
```

```python
input_file = "training.csv"

df = pd.read_csv(input_file, header = 1, names = names)
```

```python
df
```

```python
df.shape
```

```python
target = "class"
target_array = df.pop(target).values
```

```python
df.shape
```

The ***protein structure*** data set contains 200 input features (phi and psi angle pairs) and 1 output variable (the class label).

### 4.3. Assigning *input* and *output* variables
The 200 input variables to `X_train` and the output variable (class label) to `Y_train`

```python
X_train = df
Y_train = target_array
```

### 4.3. Data dimension

```python
X_train.shape
```

```python
Y_train.shape
```

## 5. Build Classification Model using Random Forest

Creating a model `clf` and giving it two variables: `X_train`, which consists of data on which the classification will be based, and `Y_train`, which consists of labels (to represent classification results).

```python
clf = RandomForestClassifier()
```

`fit` function displays the default parameters that are used for the model.

```python
clf.fit(X_train, Y_train)
```

## 6. Feature Importance

```python
print(clf.feature_importances_)
```

### 7. Validation set preparation

```python
input_file_valid = "validation.csv"
df_valid = pd.read_csv(input_file_valid, header = 1, names = names)

valid_target_array = df_valid.pop(target).values
```

```python
X_valid = df_valid
Y_valid = valid_target_array
```

```python
X_valid.shape, Y_valid.shape
```

## 8. Make prediction (using validation set)

```python
print(clf.predict(X_valid))
```

#### *Actual class labels*

```python
print(Y_valid)
```

## 9. Model Performance (for validation set)

```python
print(clf.score(X_valid, Y_valid))
```
